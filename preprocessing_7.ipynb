```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '28a4d63a51fd003a26ae5b50b3d0832d'
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie() 
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    if not result:
      return np.NaN
    else:
      movie_id = result[0].id
      response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
      data_json = response.json()
      if data_json['genres']:
          genre_str = " " 
          for i in range(0,len(data_json['genres'])):
              genres.append(data_json['genres'][i]['name'])
          return genre_str.join(genres)
      else:
          return np.NaN
```

```python
df_hin['genres'] = df_hin['Title'].map(lambda x: get_genre(str(x)))
```

```python
df_hin = df_hin.rename(columns={'Director':'director_name'})
```

```python
df_hin = df_hin.rename(columns={'Genre':'genres'})
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_hin['actor_1_name'] = df_hin['Cast'].map(lambda x: get_actor1(str(x)))
```

```python
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df_hin['actor_2_name'] = df_hin['Cast'].map(lambda x: get_actor2(str(x)))
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df_hin['actor_3_name'] = df_hin['Cast'].map(lambda x: get_actor3(str(x)))
```
